```python
movies_df = pd.read_csv('Attachments/Q3_Dataset/movies.csv')
ratings_df = pd.read_csv('Attachments/Q3_Dataset/ratings.csv')
```

```python
users_num = len(set(ratings_df['userId']))
users_num
```

```python
movies_num = len(set(movies_df['movieId']))
movies_num
```

```python
print("Shape of movies dataset:", movies_df.shape)
print("Shape of ratings dataset:", ratings_df.shape)
```

```python
rating = ratings_df['rating']
train_X = (rating-rating.min())/(rating.max()-rating.min())
ratings_df['rating'] = train_X
ratings_df.head()
```

```python
movies_df['List index'] = movies_df.index
movies_df
```

```python
joint_df = ratings_df.merge(movies_df, left_on='movieId', right_on='movieId')
joint_df
```

```python
joint_df.drop(['timestamp', 'title', 'genres', 'movieId'], axis=1, inplace=True)
joint_df.head()
```

```python
df = joint_df.groupby(by="userId")
df.get_group(1)
```

```python
class RBM():
    def __init__(self,Nv,Nh):
        self.w = np.random.random(size = (Nh,Nv))
        self.b_h = np.random.random(size = (Nh, 1))
        self.b_v = np.random.random(size = (Nv, 1))

    def sigmoid(self,x):
        return 1./(1. + np.exp(-x))

    def forward(self,x):
        z = self.w @ x + self.b_h
        a = self.sigmoid(z)
        return a

    def backward(self,a):
        z = self.w.T @ a + self.b_v
        x = self.sigmoid(z)
        return x
    
    def update(self,x,lr=9e-3):
        a = self.forward(x)
        x_prime = self.backward(a)
        a_prime = self.forward(x_prime)
        d_w = lr*(a@x.T - a_prime@x_prime.T)
        d_b_v = lr*(x - x_prime)
        d_b_h = lr*(a - a_prime)

        self.w = self.w + d_w
        self.b_v = self.b_v + d_b_v
        self.b_h = self.b_h + d_b_h

    def loss(self,x):
        a = self.forward(x)
        x_prime = self.backward(a)
        return (np.abs(x - x_prime)).mean()
    
    def predict(self,x):
        return self.backward(self.forward(x))
```

```python
epochs = 20
hidden_units = 20

model = RBM(movies_num, hidden_units)
error = []
for _ in range(epochs):
    loss = 0   
    for i in range(1, users_num+1):
        if i == 75:
            continue
        user_i = df.get_group(i)
        x = np.zeros((movies_num, 1))
        x[user_i['List index'], 0] = user_i['rating']
        model.update(x)
        loss += model.loss(x)
    print("epoch:", _+1, "\tloss:", round(loss/users_num, 5))
    error.append(loss/users_num)
```

```python
plt.plot(range(1, 21), error, color='red')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('Loss')
plt.xticks(range(1, 21))
plt.grid()
plt.show()
```

```python
user_75 = df.get_group(75)
x = np.zeros((movies_num, 1))
x[user_75['List index'], 0] = user_75['rating']

pred = model.predict(x)
ind = np.argsort(pred, axis=0)[-15:, 0]
movies_df['title'][ind]
```
